Importing necessary libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

read the image

```python
img = cv2.imread('cat.jpg')
cv2_imshow(img)
cv2.waitKey(0)
```

image to Grayscale

```python
greyscale_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(greyscale_img)
cv2.waitKey(0)
```

Grayscale into Inverted

```python
inverted_image = 255 - greyscale_img
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

Gaussian Blur

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred_image)
cv2.waitKey(0)
```

Blur Image Inverted

```python
inv_blur_img = 255 - blurred_image
cv2_imshow(inv_blur_img)
cv2.waitKey(0)
```

Sketched Form

```python
sketch = cv2.divide(greyscale_img, inv_blur_img, scale=256.0)
cv2_imshow(sketch)
cv2.waitKey(0)
```

```python
cv2_imshow(img)
cv2_imshow(sketch)
cv2.waitKey(0)
```
